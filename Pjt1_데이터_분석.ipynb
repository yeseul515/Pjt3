## 상관관계
- 'Card_Category'는 범주형 변수이므로 피어슨 상관계수가 아닌 카이제곱 검정(Chi-square test)을 사용하여 'Card_Category'와 'Attrition_Flag' 간의 상관관계를 계산하였음
- 귀무가설 : 'Card_Category'와 'Attrition_Flag' 간에는 상관관계가 없다.
- 대립가설 : 'Card_Category'와 'Attrition_Flag' 간에는 상관관계가 있다.
- p-value가 0.5 이상이라는 것은 우리가 고려한 두 변수 간의 관계가 통계적으로 유의하지 않다는  것을 의미 → p-value가 0.5 이상이면 귀무가설을 기각할 충분한 증거가 없다는 것
- 따라서 'Card_Category'와 'Attrition_Flag' 간에는 통계적으로 유의한 관련성이 없으며, 두 변수 간의 상관성은 유의하지 않음

```python
from scipy.stats import chi2_contingency

# 카이제곱 검정 실행
contingency_table = pd.crosstab(df['Card_Category'], df['Attrition_Flag'])
chi2, p, dof, expected = chi2_contingency(contingency_table)

print(f"Chi-square Statistic: {chi2}")
print(f"P-value: {p}")
```

- 범주형 변수를 인코딩하여 숫자로 변환하면 상관 계수를 계산할 수 있지만, 이러한 상관 계수는 실제로는 범주형 변수 간의 선형 관계를 나타내는 것이 아니라, 범주 간의 순서 또는 순위에 대한 상관 관계를 나타내는 것. 범주형 변수를 인코딩하면서 순서 정보를 부여하는 경우에만 의미가 있음

- 예를 들어, 성별 같은 경우에는 남자를 0, 여자를 1로 인코딩하면 순서를 나타내지 않기 때문에 상관 계수를 계산할 때 유용하지 않음. 그러나 학력 수준과 같이 범주 간에 순서가 있는 경우에는 순서 정보를 나타내기 위해 인코딩하여 상관 계수를 계산할 수 있음

- 그러나 일반적으로 범주형 변수 간의 상관 관계를 파악하기 위해서는 카이제곱 검정과 같은 비모수적인 방법이 더 적절함. 범주형 변수를 그대로 사용하여 교차표를 생성하고 이를 히트맵으로 시각화하는 것이 보다 의미 있는 결과를 제공할 수 있음

```python
cross_tab = pd.crosstab(index=[df['Card_Category'], df['고객 연령대']], columns=df['Attrition_Flag'])
cross_tab
```

```python
# 카이제곱 검정 수행
chi2, p_value, _, _ = chi2_contingency(cross_tab)

print("Chi-square statistic:", chi2)
print("p-value:", p_value)

if p_value < 0.05:
    print("p-value가 유의 수준보다 작으므로, 세 변수 간에는 통계적으로 유의미한 상관 관계가 있습니다.")
else:
    print("p-value가 유의 수준보다 크므로, 세 변수 간에는 통계적으로 유의미한 상관 관계가 없습니다.")
```

### 신용 한도 범주화
- 분포가 불균형한 수치형 변수를 범주화할 때, 일반적으로 변수를 일정한 구간으로 나누는 구간화(binning) 기법 사용. 이를 통해 데이터의 분포를 더 균일하게 만들거나 더 의미 있는 그룹으로 나눌 수 있음

- 등간격 구간화 (Equal Width Binning): 변수의 전체 범위를 동일한 길이의 구간으로 나눔. 이 방법은 분포의 균형을 유지하면서 데이터를 그룹화하지만 이상치(outliers)에 민감할 수 있음

- 등빈도 구간화 (Equal Frequency Binning): 데이터 포인트의 수가 각 구간에서 동일하도록 구간을 나눔. 이 방법은 분포의 불균형을 보다 균형있게 만들 수 있지만 일부 구간은 크기가 다를 수 있음

- 기준값을 사용한 구간화 (Quantile Binning): 변수를 주어진 백분위수 또는 분위수에 따라 나눕니다. 예를 들어, 25%, 50%, 75%의 백분위수를 기준으로 데이터를 구간화할 수 있음. 이 방법은 데이터의 분포를 고려하여 구간을 나누므로 이상치에 민감하지 않음

```python
df['신용 한도'] = 0
df
```

```python
quantiles = df['Credit_Limit'].quantile([0.25, 0.5, 0.75])
quantiles
```

```python
# 기준값을 사용한 구간화
for i in range(len(df['Credit_Limit'])):
    if df['Credit_Limit'][i] <= quantiles[0.25]:
        df['신용 한도'][i] = 'Q1'
    elif df['Credit_Limit'][i] <= quantiles[0.5]:
        df['신용 한도'][i] = 'Q2'
    elif df['Credit_Limit'][i] <= quantiles[0.75]:
        df['신용 한도'][i] = 'Q3'
    else:
        df["신용 한도"][i] = 'Q4'
```
